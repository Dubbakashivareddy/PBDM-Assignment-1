```python
import pandas as pd
import glob
import os


folderpath = 'csse_covid_19_daily_reports_us'
csv_path = os.path.join(folderpath, '*.csv')
csv_files = glob.glob(csv_path)
```

```python
target_month = int(input("Target month (MM): "))
target_year = int(input("Target year (YYYY): "))
target_dataframes = []
for file in csv_files:
    file_name = os.path.basename(file)
    file_date_str = file_name.split('.')[0]
    month, day, year = file_date_str.split('-')
    file_date = pd.to_datetime(f"{month}-{day}-{year}", format='%m-%d-%Y')
    if file_date.month == target_month and file_date.year == target_year:
        df = pd.read_csv(file)
        target_dataframes.append(df)

combined_dataframe = pd.concat(target_dataframes)
dataframe=combined_dataframe
dataframe
```
